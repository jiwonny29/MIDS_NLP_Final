# Amazon Review Data Cleaning

```python
import pandas as pd

# First Dataset 
amazon1 = pd.read_csv('C:/Users/wonny/Downloads/nlp/0_rawdata/Reviews.csv')
amazon1.head()
```

```python
amazon1.shape
```

```python
# Imbalanced Dataset 
amazon1['Score'].value_counts()
```

```python
amazon1_subset = amazon1[['Score', 'Text']]
amazon1_subset
```

```python
# Rename the column names 
amazon1_1 = amazon1_subset.copy()
amazon1_1.rename(columns={'Score' : 'rating', 'Text' : 'review'}, inplace=True)
amazon1_1
```

```python
# Second Dataset
amazon2 = pd.read_csv("C:/Users/wonny/Downloads/nlp/0_rawdata/cleaned_reviews.csv")
amazon2.head()
```

```python
amazon2.shape
```

```python
# Imbalanced dataset again
amazon2['review_score'].value_counts()
```

```python
amazon2_subset = amazon2[['review_score', 'cleaned_review']]
amazon2_subset
```

```python
amazon2_2 = amazon1_subset.copy()
amazon2_2.rename(columns={'Score' : 'rating', 'Text' : 'review'}, inplace=True)
amazon2_2
```

```python
# Third Dataset
amazon3 = pd.read_csv("c:/users/wonny/Downloads/nlp/0_rawdata/amazon_reviews.csv")
amazon3.head()
```

```python
amazon3.shape
```

```python
amazon3['overall'].value_counts()
```

```python
amazon3_subset = amazon3[['overall', 'reviewText']]
amazon3_subset
```

```python
amazon3_3 = amazon1_subset.copy()
amazon3_3.rename(columns={'Score' : 'rating', 'Text' : 'review'}, inplace=True)
amazon3_3
```

```python
result_df = pd.concat([amazon1_1, amazon2_2, amazon3_3])
result_df
```

```python
# Merge three datasets: df1, df2, df3. Then, create a new column 'sentiment'.
# Assign sentiment values based on the 'rating' column as follows:
# 1 for positive reviews (where rating is 4 or 5),
# 0 for neutral reviews (where rating is 3),
# -1 for negative reviews (where rating is 1 or 2).

result_df['sentiment'] = result_df['rating'].apply(lambda x: 1 if x in [4, 5] else (0 if x == 3 else -1 if x in [1, 2] else None))
result_df
```

```python
# Count occurrences of each sentiment value (1, 0, -1) in the 'sentiment' column.
result_df['sentiment'].value_counts()
```

```python
# Equalize the number of positive (1), neutral (0), and negative (-1) rows to 120K each.
sampled_df = result_df.groupby('sentiment').apply(lambda x: x.sample(n=120000))
sampled_df = sampled_df.reset_index(drop=True)
sampled_df
```

```python
# Rearrange the column order 
sampled_df = sampled_df[['rating', 'sentiment', 'review']]
sampled_df
```

```python
# Display a sample of the 'review' text data for inspection.
sampled_df['review'].sample(10)
```

```python
# Calculate the average word length in the 'review' column, considering that an average English sentence has 15-20 words and typically spans 5-6 sentences.
lengths = sampled_df['review'].str.len()
average_length = lengths.mean()
print(average_length)
```

```python
# Save the dataframe to a CSV file for future model training and analysis.
sampled_df.to_csv('C:/Users/wonny/Downloads/nlp/1_data_cleaning/amazon_base_data.csv', index=False)
```

```python
# Verify that the 'rating' and 'sentiment' columns do not contain any anomalous values.
print(sampled_df['rating'].unique())
print(sampled_df['sentiment'].unique())
```

# Ebay Review Data Cleaning

```python
import pandas as pd 

ebay = pd.read_csv("C:/Users/wonny/Downloads/nlp/0_rawdata/ebay_reviews.csv")
ebay_subset = ebay[['category', 'rating', 'review content']]
ebay_subset['sentiment'] = ebay_subset['rating'].apply(lambda x: 1 if x in [4, 5] else (0 if x == 3 else -1 if x in [1, 2] else None))
ebay_subset = ebay_subset[['category', 'rating', 'sentiment', 'review content']]
ebay_df = ebay_subset.groupby('sentiment').apply(lambda x: x.sample(n=1350))
ebay_df = ebay_df.reset_index(drop=True)
ebay_df
```

```python
ebay_df['sentiment'].value_counts()
```

```python
ebay_df['category'].value_counts()
```

```python
ebay_df.to_csv('C:/Users/wonny/Downloads/nlp/1_data_cleaning/ebay_base_data.csv', index=False)
```

```python
ebay_df_copy = ebay_df.copy()
ebay_upd_data = ebay_df_copy[['rating', 'sentiment', 'review content']]
ebay_upd_data.rename(columns={'review content': 'review'}, inplace=True)
ebay_upd_data
```

```python
final_df = pd.concat([sampled_df, ebay_upd_data])
final_df
```

```python
final_df['category'] = final_df['sentiment'].apply(lambda x: "positive" if x == 1 else ("neutral" if x == 0 else "negative"))
final_df = final_df[['category', 'rating', 'sentiment', 'review']]
final_df
```
